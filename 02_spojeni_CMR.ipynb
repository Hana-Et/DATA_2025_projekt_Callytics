```python
import pandas as pd
import os
```

```python
data_path=r"C:\Users\lucie\OneDrive\Desktop\Projekt\2024_01_03"
```

```python
filenames=os.listdir(data_path)
```

```python
filenames
```

```python
filenames_cmr=[]
for filename in filenames:  
    if filename[:3]=='cmr':
        filenames_cmr.append(filename)
```

```python
cmr_dataframes=[]

for filename in filenames_cmr[:]:
    file_path=data_path + '\\' + filename
    df= pd.read_csv(file_path, header=[0,1])
    cmr_dataframes.append(df)
```

```python
cmr_joined_df=pd.concat(cmr_dataframes)
```

```python
cmr_joined_df.to_csv('concat_cmd_all_files.csv')
```
